```python
disasters['Location'] = disasters['Location'].fillna('').astype(str).str.lower().str.strip().apply(unidecode.unidecode)   
rhomis_meta['Regions'] = rhomis_meta['Regions'].fillna('').astype(str).str.lower().str.strip().apply(unidecode.unidecode)
```

```python


matches = []

for country in rhomis_meta['Country full name'].unique():
    rhomis = rhomis_meta[rhomis_meta['Country full name'] == country]
    disaster_countries = disasters[disasters['Country'] == country]

    if disaster_countries.empty:   # skip if no locations for that country
        continue

    for loc in rhomis['Regions']:
        best = process.extractOne(loc, disaster_countries['Location'], scorer=fuzz.token_sort_ratio)
        if best is not None and best[1] > 65:  # similarity threshold
            matches.append((country, loc, best[0], best[1]))

```

```python
matches
```

```python
for country, rhomis_region, matched_loc, score in matches:
    print(f"\n=== Match in {country} ({score:.1f}) ===")
    print("RHoMIS region:", rhomis_region)

    # find rows in rhomis_meta
    rhomis_rows = rhomis_meta[
        (rhomis_meta['Country full name'] == country) &
        (rhomis_meta['Regions'] == rhomis_region)
    ]
    print("RHoMIS rows:")
    print(rhomis_rows)

    # find rows in disasters
    disaster_rows = disasters[disasters['Location'] == matched_loc]
    print("Disaster rows:")
    print(disaster_rows)
```
